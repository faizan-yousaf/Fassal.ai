```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import csv
# show all the columns, rows
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
url = "http://amis.pk/DistrictCities.aspx"

response = requests.get(url)

soup = BeautifulSoup(response.text, "html.parser")
```

```python
print(soup.prettify())
```

```python
# Find the div with id "master_contentplaceholder"
div = soup.find("div", id="master_contentplaceholder")


# Find the table with id "AutoNumber1" within that div
table = div.find("table", id="AutoNumber1")


# Extract the table rows
rows = table.find_all("tr")


# Loop through the rows and print the data
for row in rows:
    columns = row.find_all("td")
    for column in columns:
        print(column.get_text(strip=True))
```

```python
div = soup.find("div", id="master_contentplaceholder")

# Find the table with id "AutoNumber1" within that div
table = div.find("table", id="AutoNumber1")

# Extract all anchor tags within the table
anchors = table.find_all("a")

# Loop through and print the href and text of each anchor tag
for anchor in anchors:
    print(f"Text: {anchor.get_text(strip=True)}, URL: {anchor['href']}")
    
```

```python
# find the div with id "master_contentplaceholder"
div = soup.find("div", id="master_contentplaceholder")

# the table with id "AutoNumber1":
table = div.find("table", id="AutoNumber1")

# Extract all "a" tags:
anchors = table.find_all("a")

# Open a CSV file to write
with open("anchor_tags.csv", mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["Text", "URL"])  # Write header row
    
    # Loop through the "a" tags:
    for anchor in anchors:
        writer.writerow([anchor.get_text(strip=True), anchor['href']])

print("Anchor tags have been saved to anchor_tags.csv")
```
